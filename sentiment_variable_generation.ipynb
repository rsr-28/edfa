```python
def split_text_into_chunks(text, max_tokens=1500):
    words = text.split()
    chunks = []
    current_chunk = []

    for word in words:
        current_chunk.append(word)
        if len(current_chunk) >= max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = []

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks
```

```python
def get_sentiment_openai(text):
    #Initialize the OpenAI client
    chunks = split_text_into_chunks(text, max_tokens=1500)  # Split text into chunks
    sentiments = []
    res = 0;
    n = 0;
    for chunk in chunks:
      response = client.chat.completions.create(
          messages=[
              {
                  "role": "user",
                  "content": f"Provide a sentiment analysis for the following text:\n{chunk}",
              }
          ],
          model="gpt-3.5-turbo",
      )
      res += float(response.choices[0].message.content.strip())
      n += 1
    sentiments.append(response.choices[0].message.content.strip())  # Store the sentiment for each chunk
    return res/n  # Return sentiments for all chunks
```

```python
from collections import Counter
def get_sentiment_categorical(text):
    chunks = split_text_into_chunks(text, max_tokens=1500)
    total_score = 0
    count = 0
    sentiment_counts = Counter()
    for chunk in chunks:
        try:
            response = client.chat.completions.create(
                messages=[
                    {
                        "role": "user",
                        "content": (
                            "You are a sentiment analysis assistant. "
                            "Your task is to analyze the sentiment of the provided text."
                            "Categorize the sentiment, where 0 is negative, 1 is neutral and 2 is positive."
                            "Only provide integral numerical value:\n\n"
                            f"{chunk}"
                        ),
                    }
                ],
                model="gpt-3.5-turbo",
                temperature=0,  # Lower temperature for more deterministic output
            )
            sentiment_score = int(response.choices[0].message.content.strip())
            sentiment_counts[sentiment_score] += 1
        except Exception as e:
            print(f"Error processing chunk: {e}")
            continue

    # Return the sentiment category with the highest frequency
    if sentiment_counts:
        print(sentiment_counts.most_common(1)[0][0])
        return sentiment_counts.most_common(1)[0][0]  # Return the category with the highest frequency
    else:
        return 1  # Default to neutral if no valid responses were obtained
```
